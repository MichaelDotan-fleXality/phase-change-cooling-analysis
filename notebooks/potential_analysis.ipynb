# Phase-Change Cooling System Potential Analysis

This notebook demonstrates how to perform potential analysis for phase-change cooling systems.

## Overview

Phase-change cooling systems utilize latent heat of vaporization/condensation, providing:
- Higher energy density
- Potential for thermal storage
- Different efficiency characteristics

## Key Parameters

- **EER (Energy Efficiency Ratio)**: Cooling capacity / Electrical power input
- **Latent Heat Capacity**: Energy absorbed/released during phase change
- **Phase Change Temperature**: Temperature at which phase transition occurs

## 1. Load or Create Sample Data


```python
# Create sample data
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 1, 8)
dates = pd.date_range(start=start_date, end=end_date, freq='15min')

df = pd.DataFrame(index=dates)

# Simulate EVU meter readings (site consumption - PV)
df['EVU Meter'] = 100 + 20 * np.sin(np.arange(len(df)) * 2 * np.pi / 96)

# Simulate cooling power
df['Cooling Power'] = 50 + 10 * np.sin(np.arange(len(df)) * 2 * np.pi / 96)

# Simulate spot market prices
df['Spot Market Price (€/MWh)'] = 50 + 30 * np.sin(np.arange(len(df)) * 2 * np.pi / 96 + np.pi/4)

# Optional: PV power
hours = df.index.hour + df.index.minute / 60
df['PV Power'] = np.where(
    (hours >= 6) & (hours <= 18),
    150 * np.sin((np.pi / 12) * (hours - 6)),
    0
)

print(f"Data shape: {df.shape}")
print(f"Date range: {df.index.min()} to {df.index.max()}")
df.head()
```
